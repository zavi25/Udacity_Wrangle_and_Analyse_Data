### Cleaned Dataset Description

After cleaning the datasets and merging into one dataset, the final dataset consists of 1881 rows and 23 columns, the columns include:
- tweet_id - Unique id for each tweet.
- source - The device the tweet originated from.
- text - The text contained in the tweet.
- expanded_urls - The tweet Url's.
- rating_numerator - The rating numerator out of 10.
- rating_denominator - The rating denominator taken as 10.
- name - The dog name.
- dog_size - The classification of the dogs by size
- date - The date the tweet was made.
- time - The time of the tweet
- jpg_url - The dog image url
- img_num - The image number
- p1 - The algorithm's #1 prediction
- p1_conf - Accuracy confindence for the #1 prediction
- p1_dog -  Whether or not the #1 prediction is a breed of dog
- p2 - The algorithm's #2 prediction
- p2_conf -Accuracy confindence for the #2 prediction
- p2_dog -  Whether or not the #2 prediction is a breed of dog
- p3 - The algorithm's #3 prediction
- p3_conf -  Accuracy confindence for the #3 prediction
- p3_dog - Whether or not the #3 prediction is a breed of dog
- retweet_count - The number of retweets per tweet
- favorite_count - The number of likes per tweet
